*Problem 2*

The mean values of the columns `Flavanoids` and `Total_phenols` in the wine dataset are 2.03 and 2.29, respectively. Although these means are nearby, does this imply that the distributions of these two variables are practically equivalent?

To answer this, create a histogram and overlay the probability density curve for each of the 2 variables. Discuss your findings, particularly focusing on the shape and spread of the distributions. Do add a note on the peaks of the histogram. (2 points)



````r

data <- read.csv("/kaggle/input/wine-quality/wine_quality_combined.csv", header = TRUE, stringsAsFactors = FALSE)
wine_data <- as.data.frame(data)
wine_data
````

````r

library(ggplot2)


ggplot(wine_data, aes(x = Flavanoids)) +
  geom_histogram(aes(y = ..density..), bins = 30, fill = "skyblue", color = "black", alpha = 0.7) +
  geom_density(alpha = 0.2, fill = "blue") +
  labs(title = "Histogram and Density Curve of Flavanoids",
       x = "Flavanoids",
       y = "Density") +
  theme_minimal()

ggplot(wine_data, aes(x = Total_phenols)) +
  geom_histogram(aes(y = ..density..), bins = 30, fill = "salmon", color = "black", alpha = 0.7) +
  geom_density(alpha = 0.2, fill = "red") +
  labs(title = "Histogram and Density Curve of Total Phenols",
       x = "Total Phenols",
       y = "Density") +
  theme_minimal()
````

Despite similar mean values the distributions of these variables differ significantly.

Flavanoids: The distribution is multimodal with a right skew, showing peaks around 1, 2, and 3, and a wider spread.

Total Phenols: The distribution is more symmetric and concentrated, with two main peaks around 2 and 3.

The close means do not imply similar distributions, as shown by the differences in shape, spread, and peak patterns.

*Problem 4*

Explain Dimensionality Reduction. Perform PCA on the dataset and extract the proportion of variance explained by each principal component. How many principal components should be retained based on the `scree plot`? Examine the loadings of the first two principal components. Which variables contribute most to these components? (2 points)

**Hint**:
Scree Plot is a common method for determining the number of PCs to be retained through a graphical representation. A Scree Plot is a simple line segment plot that shows the eigenvalues for each individual PC. 

You can learn more about a scree plot [here](https://sanchitamangale12.medium.com/scree-plot-733ed72c8608)

**Hint**:
In PCA, the contribution of each feature to a principal component is called the `loading`. Loadings are compared with the absolute value



Dimensionality Reduction : process of removing unimportant attributes

````r

library(stats)

wine_pca <- prcomp(wine_data[, sapply(wine_data, is.numeric)], scale. = TRUE)

summary(wine_pca)
````

````r

explained_variance <- summary(wine_pca)$importance[2, ]

explained_variance
````

````r

plot(wine_pca, type = "l", main = "Scree Plot")
````

We should retain 2 principal components based on the scree plot

````r

loadings <- wine_pca$rotation[, 1:2]

loadings
````

PC1: The variables "Flavanoids," "Total_phenols," "X0D280_0D315_of_diluted_wines," and "Color_intensity" have the most significant contributions.
PC2: The variables "Color_intensity," "Alcohol," "Proline," and "Ash" have the most significant contributions.

*Problem 5* 

Now, for the selected components, find the percentage contribution of each feature to each of the components. (Hint: First find the square of the loadings also known as the cos2 values - find percentage contribution using these values) (2 points)

````r

loadings <- wine_pca$rotation[, 1:2]  

squared_loadings <- loadings^2

sum_squared_loadings <- colSums(squared_loadings)

percentage_contributions <- sweep(squared_loadings, 2, sum_squared_loadings, FUN = "/") * 100

percentage_contributions_df <- as.data.frame(percentage_contributions)
colnames(percentage_contributions_df) <- c("PC1_percentage", "PC2_percentage")

print(percentage_contributions_df)
````
